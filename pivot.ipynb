```python
import pandas as pd
df = pd.read_excel('Pivot.xlsx')
df.head()
```

```python
df.pivot_table(values='Sprzedaż',index='Przedstawiciel',columns='Region',aggfunc='sum').fillna(0).round(2)
```

```python
df.pivot_table(values='Sprzedaż',index=['Region','Przedstawiciel'],aggfunc='sum').round(2)
```

```python
df.pivot_table(values='Sprzedaż',index='Region',aggfunc=['count', 'max', 'min']).round(0)
```

```python
def commission_fee(x):


    if x <= 300:
        return 0
    elif x <= 900:
        return x * 0.03
    else:
        return x * 0.06
```

```python
df['commission_fee'] = df['Sprzedaż'].apply(lambda x: commission_fee(x))
```

```python
df['Produkt_len'] = df['Produkt'].apply(len)

df.head()
```

```python
import math
df['#_opakowań'] = df['Sztuki'].apply(lambda x: math.ceil(x/5))
df.head()
```

```python
def bonus(row):

    margin = (row['Sprzedaż'] - row['Koszty'])/row['Sprzedaż']

    if margin > 0.55:
        return 200
    else:
        return 0
```

```python
df['Bonus'] = df.apply(lambda row: bonus(row),axis=1)
df.sample(10)
```

```python
car_dict = dict(zip(df['Przedstawiciel'].unique(),['Mazda','Toyota','BMW','Audi','Fiat','Seat']))
car_dict
```

```python
df['Marka_samochodu'] = df['Przedstawiciel'].map(car_dict)
df.sample(10)
```

```python
df = df.apply(lambda col: col.str.upper() if col.dtypes == 'object' else col)
df
```

```python
df['Data'] = pd.to_datetime(df['Data'])
df['Data'].dt.day_name().head()
```
