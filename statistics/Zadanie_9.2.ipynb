```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, mode

# Generowanie danych z rozkadu normalnego
data_100 = np.random.normal(3, 1, 100)
data_10000 = np.random.normal(3, 1, 10000)

# Generowanie osi x do wizualizacji PDF
x = np.linspace(-5, 10, 10000)

# Rysowanie wykresw PDF dla dwch zestaww danych
plt.figure(figsize=(12, 8))
plt.plot(x, norm.pdf(x, 3, 1), label='Rozkad normalny (100 prbek, =3, =1)')
plt.plot(x, norm.pdf(x, 3, 1), label='Rozkad normalny (10,000 prbek, =3, =1)', linestyle='dashed')
plt.title("Rozkady normalne dla dwch prb (100 i 10,000 prbek)")
plt.xlabel("x")
plt.ylabel("Prawdopodobiestwo")
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Obliczanie statystyk dla data_100
mean_100 = np.mean(data_100)
median_100 = np.median(data_100)
mode_100 = mode(data_100)[0]
q0_100 = np.min(data_100)
q1_100 = np.percentile(data_100, 25)
q3_100 = np.percentile(data_100, 75)
q4_100 = np.max(data_100)

# Obliczanie statystyk dla data_10000
mean_10000 = np.mean(data_10000)
median_10000 = np.median(data_10000)
mode_10000 = mode(data_10000)[0]
q0_10000 = np.min(data_10000)
q1_10000 = np.percentile(data_10000, 25)
q3_10000 = np.percentile(data_10000, 75)
q4_10000 = np.max(data_10000)

# Rysowanie histogramw
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Histogram dla data_100
axes[0].hist(data_100, bins=20, color='lightblue', edgecolor='black', alpha=0.7)
axes[0].axvline(mean_100, color='red', linestyle='-', linewidth=2, label=f'rednia: {mean_100:.2f}')
axes[0].axvline(median_100, color='green', linestyle='--', linewidth=2, label=f'Mediana: {median_100:.2f}')
axes[0].axvline(mode_100, color='blue', linestyle=':', linewidth=2, label=f'Moda: {mode_100:.2f}')
axes[0].axvline(q0_100, color='purple', linestyle='-', linewidth=1, label=f'Q0: {q0_100:.2f}')
axes[0].axvline(q1_100, color='purple', linestyle='--', linewidth=1, label=f'Q1: {q1_100:.2f}')
axes[0].axvline(q3_100, color='purple', linestyle='--', linewidth=1, label=f'Q3: {q3_100:.2f}')
axes[0].axvline(q4_100, color='purple', linestyle='-', linewidth=1, label=f'Q4: {q4_100:.2f}')
axes[0].set_title("Histogram dla 100 prbek")
axes[0].set_xlabel("Wartoci")
axes[0].set_ylabel("Czsto")
axes[0].legend()
axes[0].grid(True)
axes[0].set_xticks(range(int(q0_100), int(q4_100) + 1))

# Histogram dla data_10000
axes[1].hist(data_10000, bins=50, color='lightgreen', edgecolor='black', alpha=0.7)
axes[1].axvline(mean_10000, color='red', linestyle='-', linewidth=2, label=f'rednia: {mean_10000:.2f}')
axes[1].axvline(median_10000, color='green', linestyle='--', linewidth=2, label=f'Mediana: {median_10000:.2f}')
axes[1].axvline(mode_10000, color='blue', linestyle=':', linewidth=2, label=f'Moda: {mode_10000:.2f}')
axes[1].axvline(q0_10000, color='purple', linestyle='-', linewidth=1, label=f'Q0: {q0_10000:.2f}')
axes[1].axvline(q1_10000, color='purple', linestyle='--', linewidth=1, label=f'Q1: {q1_10000:.2f}')
axes[1].axvline(q3_10000, color='purple', linestyle='--', linewidth=1, label=f'Q3: {q3_10000:.2f}')
axes[1].axvline(q4_10000, color='purple', linestyle='-', linewidth=1, label=f'Q4: {q4_10000:.2f}')
axes[1].set_title("Histogram dla 10,000 prbek")
axes[1].set_xlabel("Wartoci")
axes[1].legend()
axes[1].grid(True)
axes[1].set_xticks(range(int(q0_10000), int(q4_10000) + 1))

# Dopasowanie osi Y dla kadego histogramu osobno
axes[0].set_ylim(0, axes[0].get_ylim()[1])
axes[1].set_ylim(0, axes[1].get_ylim()[1])

# Wywietlanie wynikw
plt.tight_layout()
plt.show()

print("Statystyki dla 100 prbek:")
print(f"rednia: {mean_100:.2f}, Mediana: {median_100:.2f}, Moda: {mode_100:.2f}")
print(f"Q0: {q0_100:.2f}, Q1: {q1_100:.2f}, Q3: {q3_100:.2f}, Q4: {q4_100:.2f}")

print("\nStatystyki dla 10,000 prbek:")
print(f"rednia: {mean_10000:.2f}, Mediana: {median_10000:.2f}, Moda: {mode_10000:.2f}")
print(f"Q0: {q0_10000:.2f}, Q1: {q1_10000:.2f}, Q3: {q3_10000:.2f}, Q4: {q4_10000:.2f}")
```

##### Obserwacje: W przypadku rozkadu normalnego linie pokrywaj si dla obu prbek. Jeli chodzi o statystyki to im wicej prbek, tym bliej zmiennej miu (3) jest rednia i mediana. Nim wicej prbek tym szersze s kwartyle. Zmienne s tak zrnicowane, e trudno wyliczy mode nawet przy 10000 prbek. Naleaoby najpierw zaokrgli wyniki prbek do penych liczb ale to przy redniej 3 nie pokazaoby na histogramie jak nierwny jest on dla maej iloci prbek i jak bardzo przypominajcy PDF ju przy 10000 prbek.
