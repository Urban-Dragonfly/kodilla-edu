```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_wine
```

```python
df = pd.DataFrame(load_wine().data, columns=load_wine().feature_names)
df.head()
```

```python
df.describe()
```

```python
print(df.shape)
```

```python
df.isnull().sum()
```

```python
df_2features = df[['alcohol', 'color_intensity']]
```

```python
display(df_2features.head())
```

```python
df_2features.corr()
```

```python
fig, ax = plt.subplots()
ax.scatter(df_2features['alcohol'], df_2features['color_intensity'])
ax.set_xlabel('Alcohol')
ax.set_ylabel('Color Intensity')
plt.show();
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5).fit(df_2features)

print(kmeans.labels_)
print(kmeans.cluster_centers_)
print(kmeans.inertia_)
print(kmeans.n_iter_)
print(kmeans.max_iter)

fig, ax = plt.subplots()
ax.scatter(df_2features['alcohol'], df_2features['color_intensity'], c=kmeans.labels_)

centers = kmeans.cluster_centers_
ax.scatter(centers[:, 0], centers[:, 1], c='red', s=50)

ax.set_xlabel('Alcohol')
ax.set_ylabel('Color Intensity')
plt.show();
```

```python
import numpy as np

from sklearn.datasets import make_blobs

# rodki naszych klastrw

centroids = np.array([
    [ 0.8, 2.0],
    [-0.5, 2.0],
    [-2.0, 2.0],
    [-2.5, 2.5],
    [-2.5, 1.0]
])

# wprowadzenie szumu do naszych klastrw, aby rozrzuci prbki
blob_std = np.array([0.4, 0.3, 0.1, 0.1, 0.1])

# stworzenie zbioru danych
X, y = make_blobs(
    n_samples=3000,
    centers=centroids,
    cluster_std=blob_std,
    random_state=7
)

# wykres
fig, ax = plt.subplots()
ax.scatter(X[:, 0], X[:, 1], c=y, s=5)
ax.set_xlabel('X0')
ax.set_ylabel('X1')
plt.show();
```

```python
from sklearn.cluster import KMeans

clf = KMeans(n_clusters=5, random_state=7)

# moemy uy metod fit(), predict()
clf.fit(X)
y_pred = clf.predict(X)

# albo metody fit)predict, ktra czy dwie powysze
y_pred = clf.fit_predict(X)

fig, ax = plt.subplots()
ax.scatter(X[:, 0], X[:, 1], c=y_pred, s=5)
ax.set_xlabel('X0')
ax.set_ylabel('X1')

centers = clf.cluster_centers_
ax.scatter(centers[:, 0], centers[:, 1], c='red', s=50)

plt.show();
```

```python


plt.figure(figsize=(10, 5))

def plot_decision_boundaries(model, X, resolution=0.02):
	x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
	y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
	xx, yy = np.meshgrid(np.arange(x_min, x_max, resolution),
						 np.arange(y_min, y_max, resolution))
	Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
	Z = Z.reshape(xx.shape)
	plt.contourf(xx, yy, Z, alpha=0.3)
	plt.scatter(X[:, 0], X[:, 1], c=model.predict(X), s=5)

plot_decision_boundaries(clf, X)
plt.show()

print(clf.cluster_centers_)
```
